```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
from sklearn.datasets import load_diabetes
```

The Diabetes dataset contains 10 physiological features such as age, sex, body mass index (bmi), blood pressure, and six blood serum measurements obtained for 442 diabetes patients. The target represents a quantitative measure of disease progression.

```python
load_diabetes()
```

```python
pd.DataFrame(load_diabetes().data)
```

```python
dataset=load_diabetes()

X= pd.DataFrame(data=dataset.data,
                          columns=dataset.feature_names)
y = dataset.target

# dataset.head()
```

```python
# dataset.shape
```

```python
# X=dataset # Independent features
# y=dataset.target # Dependent features
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

Standardization of datasets is a common requirement for many machine learning estimators; they might behave badly if the individual features do not more or less look like standard normally distributed data: Gaussian with zero mean and unit variance.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
## reversing the transformation
# scaler.inverse_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
# cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
np.mean(mse)
```

```python
# prediction
reg_pred=regression.predict(X_test)
reg_pred
```

R² (R-SQUARED) SCORE (or COEFFICIENT OF DETERMINATION): A key regression metric showing how much of the dependent variable's variance is explained by the model's independent variables, ranging from 0 (no fit) to 1 (perfect fit)

R² measures the proportion of variance in the target variable explained by the regression model relative to a mean-based baseline.

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
score
```

(Using 10 physiological features, our linear regression explains ~48% of the variability in disease progression across patients.)

Calculating Train MSE & Test MSE: We train the model once on training data, then use the same model to predict on both training and test sets to compute Train MSE and Test MSE.

```python
from sklearn.metrics import mean_squared_error
y_train_pred=regression.predict(X_train) # train predictions
y_test_pred=regression.predict(X_test) # test predictions

train_mse=mean_squared_error(y_train,y_train_pred)
test_mse=mean_squared_error(y_test,y_test_pred)
```
